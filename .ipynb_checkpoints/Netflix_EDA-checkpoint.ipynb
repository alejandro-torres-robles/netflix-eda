# Exploración de Datos de Netflix

import pandas as pd

# Carga el dataset
df = pd.read_csv("data/netflix_titles.csv")  
df.head()


```python
## 2. ¿Cuántas películas y series hay?
df['type'].value_counts()
```

```python
## 3. Top 5 países con más títulos
df['country'].value_counts().head(5)
```

```python
## 4. Año con más lanzamientos
df['release_year'].value_counts().head(1)
```

```python
## 5. Porcentaje de clasificación TV-MA
total = df.shape[0]  # Número total de registros
tv_ma = df[df['rating'] == 'TV-MA'].shape[0]  # Número de TV-MA
percentage = (tv_ma / total) * 100
print(f"Porcentaje de TV-MA: {percentage:.2f}%")
```

```python
## 6. Director con más títulos
df['director'].value_counts().head(1)
```

```python
## 7. Top 10 géneros más comunes (Gráfica)
import matplotlib.pyplot as plt
import seaborn as sns

# Conteo de géneros
genre_counts = df['listed_in'].value_counts().head(10)

# Gráfica
plt.figure(figsize=(12,6))
sns.barplot(x=genre_counts.values, y=genre_counts.index)
plt.title('Top 10 géneros en Netflix')
plt.xlabel('Número de títulos')
plt.ylabel('Género')
plt.show()
```
